#problem 3 week 1

```
def pass_fail(score):
    if score >= 60:
        return "Pass"
    else:
        return "Fail"


scores = [45, 67, 82, 59]

for s in scores:
    print(f"Score: {s} → {pass_fail(s)}")
```

#problem 4 week 1

```
sizes = [500, 800, 1000, 1200]
prices = [50, 80, 100, 130]


for size, price in zip(sizes, prices):
    print(f"House size: {size} sq ft → Price: {price}")


positive_trend = True

for i in range(1, len(prices)):
    if prices[i] <= prices[i - 1]:
        positive_trend = False
        break

if positive_trend:
    print("Positive trend")
```

#problem 5 week 1

```
study_hours = [2, 4, 6, 8]
scores = [55, 65, 75, 85]

increases = []
for i in range(1, len(scores)):
    increases.append(scores[i] - scores[i - 1])

average_increase = sum(increases) / len(increases)

predicted_score = scores[-1] + average_increase

print(f"Predicted score for 10 hours of study: {predicted_score}")
```

#problem 6 week 1

```
scores = [35, 58, 62, 74, 89, 45]

for score in scores:
    if score < 50:
        category = "Low"
    elif 50 <= score <= 70:
        category = "Medium"
    else:
        category = "High"

    print(f"Score: {score} → Category: {category}")
```

#problem 7 week 1

```
ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for age, height in zip(ages, heights):
    print(f"Age: {age} years → Height: {height} cm")

average_age = sum(ages) / len(ages)
print(f"\nAverage age: {average_age}")

average_height = sum(heights) / len(heights)
print(f"Average height: {average_height}")

print("Height appears to increase as age increases in this dataset.")
```
